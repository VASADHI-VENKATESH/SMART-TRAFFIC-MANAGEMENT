```python
!pip install gradio
import gradio as gr
from PIL import Image, ImageDraw, ImageFont
import io
import base64
from inference_sdk import InferenceHTTPClient
import random
import numpy as np
import time  # Import the time module

CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="CE2svnwHuufArJYxbNoL"
)

# Define consistent colors for each vehicle class
COLOR_MAP = {
    "car": (255, 0, 0),        # Red
    "bus": (0, 255, 0),        # Green
    "truck": (0, 0, 255),      # Blue
    "motorbike": (255, 255, 0), # Yellow
    "bicycle": (255, 0, 255),  # Magenta
    "person": (0, 255, 255),   # Cyan
    "auto": (255, 165, 0),     # Orange
    "van": (128, 0, 128),      # Purple
    "default": (200, 200, 200) # Grey for any undefined class
}

# Road names
ROAD_NAMES = {
    "north": "North Road",
    "south": "South Road",
    "east": "East Road",
    "west": "West Road",
    "intersection": "Intersection"
}

# Vehicle weights
VEHICLE_WEIGHTS = {
    "Car": 1,
    "Bus": 2,
    "Truck": 2,
    "Bike": 0.5,
    "Bicycle": 0.5
}

# Default signal timings (seconds)
DEFAULT_SIGNAL_TIMINGS = {
    "north": 30,
    "south": 30,
    "east": 30,
    "west": 30,
    "intersection": 0  # Reduced time for intersection
}

def determine_road_location(x, y, img_width, img_height):
    """
    Determine which road a vehicle is on based on its coordinates
    Given the 4-way junction with:
    - Vertical road width is 20% of image width
    - Horizontal road width is 25% of image height
    """
    # Calculate road boundaries
    vertical_road_start = img_width * 0.4  # 40% from left edge
    vertical_road_end = img_width * 0.6    # 60% from left edge (20% width)

    horizontal_road_start = img_height * 0.375  # 37.5% from top edge
    horizontal_road_end = img_height * 0.625    # 62.5% from top edge (25% height)

    # Check if in intersection (both on vertical and horizontal road)
    if (vertical_road_start <= x <= vertical_road_end and
        horizontal_road_start <= y <= horizontal_road_end):
        return "intersection"

    # Check if on vertical road
    elif vertical_road_start <= x <= vertical_road_end:
        if y < horizontal_road_start:
            return "north"  # North road
        else:
            return "south"  # South road

    # Check if on horizontal road
    elif horizontal_road_start <= y <= horizontal_road_end:
        if x < vertical_road_start:
            return "west"   # West road
        else:
            return "east"   # East road

    # If not on any road, determine which quadrant
    else:
        if x < vertical_road_start:
            if y < horizontal_road_start:
                return "northwest"
            else:
                return "southwest"
        else:
            if y < horizontal_road_start:
                return "northeast"
            else:
                return "southeast"

''' example data:
north {'Bus': 3, 'Car': 4, 'Truck': 1}
south {'Truck': 2, 'Car': 2}
east {'Bike': 3, 'Car': 4, 'Truck': 2}
west {'Car': 1}

# Vehicle weights
VEHICLE_WEIGHTS = {
    "car": 1,
    "bus": 2,
    "truck": 2,
    "motorbike": 0.5,
    "bicycle": 0.5
}

'''
def calculate_priority_score(vehicle_counts):
    """
    Calculate priority score based on weighted vehicle counts
    """
    priority_score = 0
    for vehicle_type, count in vehicle_counts.items():
        if vehicle_type in VEHICLE_WEIGHTS:
            priority_score += VEHICLE_WEIGHTS[vehicle_type] * count
    print(priority_score)
    return priority_score

def determine_signal_timing(priority_score):
    """
    Determine signal timing based on priority score
    """
    if priority_score <= 10:
        return 20
    elif priority_score <= 20:
        return 30
    elif priority_score <= 30:
        return 40
    else:
        return 50



"""def simulate_traffic_signals(road_counts):
    # start of the old signal function
    #Simulate traffic signal changes based on road-wise vehicle counts

    signal_timings = {}
    for road, vehicle_types in road_counts.items():
        print(road, vehicle_types)
        priority_score = calculate_priority_score(vehicle_types)
        print('---', vehicle_types, priority_score)
        signal_timing = determine_signal_timing(priority_score)
        signal_timings[road] = signal_timing
    print('signal timings \n', signal_timings)
    # Implement signal timing logic here
    # This is a simplified example - real-world scenarios need more complex logic
    roads = ["north", "south", "east", "west", "intersection"]
    # while True:
    print('--------------------------------------')
    for road in roads:
          # Default to a minimum time, ensure safety
          green_time = max(signal_timings.get(road, DEFAULT_SIGNAL_TIMINGS.get(road, 15)), 15)
          print(f"Changing {ROAD_NAMES[road]} signal to green for {green_time} seconds...")
          time.sleep(green_time)  # Simulate green light
          print(f"Changing {ROAD_NAMES[road]} signal to red...")
          time.sleep(5) # Simulate inter-green time before next signal
      # Optionally add delay between cycles
      # print('delay between cycle')
      # time.sleep(2)"""#look here for the old signal process


def simulate_traffic_signals(road_counts):
    """
    Generate traffic signal timings with white text display
    """
    # Calculate signal timings using your original method
    signal_timings = {}
    roads = ["north", "south", "east", "west", "intersection"]

    for road in roads:
        vehicle_types = road_counts.get(road, {})
        priority_score = calculate_priority_score(vehicle_types)
        green_time = determine_signal_timing(priority_score)
        signal_timings[road] = max(green_time, 15)  # Minimum 15 seconds

    # Generate HTML output with white text
    html_output = """
    <div style='
        margin: 10px;
        padding: 15px;
        background: #222;
        border-radius: 8px;
        color: white;  /* Base text color */
    '>
        <h3 style='color: white; margin-top: 0; border-bottom: 1px solid #444; padding-bottom: 8px;'>
            <i class='fas fa-traffic-light' style='color: #4CAF50; margin-right: 8px;'></i>
            Traffic Signal Sequence
        </h3>
        <div style='margin-bottom: 15px; color: #ccc; font-size: 0.9em;'>
            Each signal gets green time based on traffic, with 5s transition period
        </div>
    """

    # Calculate and display timings for each road
    for road in roads:
        green_time = max(signal_timings.get(road, DEFAULT_SIGNAL_TIMINGS.get(road, 15)), 15)
        red_time = sum(
            max(signal_timings.get(r, DEFAULT_SIGNAL_TIMINGS.get(r, 15)), 15)
            for r in roads if r != road
        ) + (len(roads)-1)*5

        html_output += f"""
        <div style='
            margin: 12px 0;
            padding: 12px;
            background: #2a2a2a;
            border-radius: 6px;
            color: white;
            border-left: 4px solid #4CAF50;
        '>
            <h4 style='margin: 0 0 8px 0; color: white;'>{ROAD_NAMES[road]}</h4>

            <div style='display: flex; align-items: center; color: white;'>
                <div style='
                    width: 14px;
                    height: 14px;
                    background: #4CAF50;
                    border-radius: 50%;
                    margin-right: 10px;
                    border: 2px solid white;
                '></div>
                <span style='color: white;'>
                    <strong>Green:</strong> {green_time}s <span style='color: #aaa; font-size: 0.9em;'>(active duration)</span>
                </span>
            </div>

            <div style='display: flex; align-items: center; margin-top: 8px; color: white;'>
                <div style='
                    width: 14px;
                    height: 14px;
                    background: #F44336;
                    border-radius: 50%;
                    margin-right: 10px;
                    border: 2px solid white;
                '></div>
                <span style='color: white;'>
                    <strong>Red:</strong> {red_time}s <span style='color: #aaa; font-size: 0.9em;'>(combined other signals)</span>
                </span>
            </div>
        </div>
        """

    # Footer with cycle information
    html_output += """
        <div style='
            margin-top: 20px;
            padding: 12px;
            background: #2a2a2a;
            border-radius: 6px;
            color: white;
            border-left: 4px solid #FFC107;
        '>
            <div style='display: flex; align-items: center; margin-bottom: 8px; color: white;'>
                <div style='
                    width: 14px;
                    height: 14px;
                    background: #FFC107;
                    border-radius: 50%;
                    margin-right: 10px;
                    border: 2px solid white;
                '></div>
                <span style='color: white;'>Transition period: <strong>5 seconds</strong> (all-red phase)</span>
            </div>

            <div style='display: flex; align-items: center; color: white;'>
                <div style='
                    width: 14px;
                    height: 14px;
                    background: #9E9E9E;
                    border-radius: 50%;
                    margin-right: 10px;
                    border: 2px solid white;
                '></div>
                <span style='color: white;'>Minimum green time: <strong>15 seconds</strong> (safety requirement)</span>
            </div>
        </div>
    </div>
    """

    return html_output





def predict(image_path):
    try:
        result = CLIENT.infer(image_path, model_id="simtrafficview5k/2")

        # Process the result and create an image with bounding boxes
        image = Image.open(image_path)
        img_width, img_height = image.size
        predictions = result['predictions']

        # Initialize road-wise vehicle counts
        road_counts = {
            "north": {},
            "south": {},
            "east": {},
            "west": {},
            "intersection": {}
        }

        # Draw boxes with different colors based on vehicle type
        for prediction in predictions:
            x = prediction['x']
            y = prediction['y']
            x1 = x - prediction['width']/2
            y1 = y - prediction['height']/2
            x2 = x + prediction['width']/2
            y2 = y + prediction['height']/2

            class_name = prediction['class']
            confidence = prediction['confidence']

            # Determine which road the vehicle is on
            road = determine_road_location(x, y, img_width, img_height)

            # Only count vehicles on actual roads (not in quadrants)
            if road in road_counts:
                # Update count for this road and vehicle type
                if class_name in road_counts[road]:
                    road_counts[road][class_name] += 1
                else:
                    road_counts[road][class_name] = 1

            # Get color for this class
            color = COLOR_MAP.get(class_name.lower(), COLOR_MAP["default"])

            # Draw bounding box and label
            draw = ImageDraw.Draw(image)
            draw.rectangle([x1, y1, x2, y2], outline=color, width=1)

            # Create label with class name, confidence, and road
            if road in road_counts:
                label = f"{class_name} ({confidence:.2f}) - {ROAD_NAMES[road]}"
            else:
                label = f"{class_name} ({confidence:.2f})"

            # # Draw text background for better visibility
            # text_width, text_height = draw.textsize(label) if hasattr(draw, 'textsize') else (len(label) * 6, 15)
            # draw.rectangle([x1, y1 - text_height - 4, x1 + text_width + 2, y1], fill=color)

            # # Draw text in white for better contrast
            # draw.text((x1 + 2, y1 - text_height - 2), label, fill=(255, 255, 255))

        # Visualize the road areas on the image
        overlay = Image.new('RGBA', image.size, (0, 0, 0, 0))
        overlay_draw = ImageDraw.Draw(overlay)

        # Draw vertical road
        overlay_draw.rectangle([
            img_width * 0.4, 0,
            img_width * 0.6, img_height
        ], fill=(100, 100, 255, 50))

        # Draw horizontal road
        overlay_draw.rectangle([
            0, img_height * 0.375,
            img_width, img_height * 0.625
        ], fill=(100, 255, 100, 50))

        # Draw intersection
        overlay_draw.rectangle([
            img_width * 0.4, img_height * 0.375,
            img_width * 0.6, img_height * 0.625
        ], fill=(255, 100, 100, 70))

        # Add road labels
        font_size = max(14, min(img_width, img_height) // 30)
        overlay_draw.text((img_width * 0.5, img_height * 0.1), "North Road", fill=(0, 0, 0, 200), anchor="ms")
        overlay_draw.text((img_width * 0.5, img_height * 0.9), "South Road", fill=(0, 0, 0, 200), anchor="ms")
        overlay_draw.text((img_width * 0.1, img_height * 0.5), "West Road", fill=(0, 0, 0, 200), anchor="ms")
        overlay_draw.text((img_width * 0.9, img_height * 0.5), "East Road", fill=(0, 0, 0, 200), anchor="ms")

        # Composite the overlay on the original image
        image = Image.alpha_composite(image.convert('RGBA'), overlay).convert('RGB')

        # Create a summary of road-wise vehicle counts
        #old count display of the project start
        """count_summary = "<div style='margin: 10px; padding: 10px; background-color: #000000; border-radius: 5px;'>"
        count_summary += "<h3>Road-wise Vehicle Counts:</h3>"

        total_vehicles = 0
        for road_name, vehicle_types in road_counts.items():
            if not vehicle_types:  # Skip roads with no vehicles
                continue

            road_total = sum(vehicle_types.values())
            total_vehicles += road_total

            count_summary += f"<h4>{ROAD_NAMES[road_name]} ({road_total} vehicles):</h4>"
            count_summary += "<ul style='list-style-type: none; padding: 0; margin-left: 20px;'>"

            for vehicle_type, count in sorted(vehicle_types.items()):
                color = COLOR_MAP.get(vehicle_type.lower(), COLOR_MAP["default"])
                color_hex = "#{:02x}{:02x}{:02x}".format(*color)
                count_summary += f"<li style='margin: 3px 0;'><span style='display: inline-block; width: 15px; height: 15px; background-color: {color_hex}; margin-right: 10px;'></span><b>{vehicle_type}:</b> {count}</li>"

            count_summary += "</ul>"

        count_summary += f"<h4>Total vehicles in all roads: {total_vehicles}</h4>"
        count_summary += "</div>" """
        #old count display end

        #new display for vehicles counter start

        # Main container with dark background and ALL text forced to white
        count_summary = "<div style='margin:5px;padding:10px;background:#222;border-radius:6px;color:white;font-family:Arial;'>"

        # Header section - ALL text white
        count_summary += "<h3 style='color:white;margin:0 0 8px 0;padding-bottom:5px;border-bottom:1px solid #444;font-size:1.1em;display:flex;align-items:center;'>"
        count_summary += "<span style='display:inline-block;width:18px;height:18px;background:#4CAF50;border-radius:50%;margin-right:8px;text-align:center;line-height:18px;font-size:12px;color:white;'>🚗</span>"
        count_summary += "Vehicle Counts</h3>"

        total_vehicles = 0

        for road_name, vehicle_types in road_counts.items():
            if not vehicle_types:
                continue

            road_total = sum(vehicle_types.values())
            total_vehicles += road_total

            # Road section - ALL text explicitly white
            count_summary += f"<div style='margin-bottom:8px;background:#2a2a2a;border-radius:4px;color:white;'>"

            # Road header - forced white text
            count_summary += f"<div style='padding:6px 10px;background:#2d2d2d;display:flex;justify-content:space-between;align-items:center;font-size:0.95em;color:white;'>"
            count_summary += f"<span style='font-weight:bold;color:white;'>{ROAD_NAMES[road_name]}</span>"  # Explicit white
            count_summary += f"<span style='background:#4CAF50;color:white;padding:2px 6px;border-radius:10px;font-size:0.8em;'>" + str(road_total) + "</span>"
            count_summary += "</div>"

            # Vehicle list - forced white text
            count_summary += "<ul style='list-style-type:none;padding:5px 10px;margin:0;color:white;'>"

            for vehicle_type, count in sorted(vehicle_types.items()):
                color = COLOR_MAP.get(vehicle_type.lower(), COLOR_MAP["default"])
                color_hex = "#{:02x}{:02x}{:02x}".format(*color)

                # Vehicle item - ALL text forced white
                count_summary += f"<li style='margin:4px 0;padding:4px 6px;display:flex;align-items:center;font-size:0.9em;color:white;'>"
                count_summary += f"<span style='display:inline-block;width:12px;height:12px;background-color:{color_hex};border-radius:50%;margin-right:8px;border:1px solid white;'></span>"
                count_summary += f"<span style='flex:1;color:white;'>{vehicle_type}</span>"  # Explicit white
                count_summary += f"<span style='background:#3a3a3a;color:white;padding:1px 6px;border-radius:8px;'>{count}</span>"
                count_summary += "</li>"

            count_summary += "</ul></div>"

        # Footer - ALL text white
        count_summary += f"<div style='margin-top:10px;padding:6px;background:#2a2a2a;border-radius:4px;text-align:center;font-size:0.95em;color:white;'>"
        count_summary += f"Total: <span style='color:#4CAF50;font-weight:bold;'>{total_vehicles}</span>"
        count_summary += "</div></div>"
        #new vehicles counter display end


        # Convert the PIL Image to a format that Gradio can display
        buffered = io.BytesIO()
        image.save(buffered, format="JPEG")
        img_str = base64.b64encode(buffered.getvalue()).decode()

        # Start traffic signal simulation in a separate thread
        # In a real-world application, this would run continuously
        #import threading
        #threading.Thread(target=simulate_traffic_signals, args=(road_counts,), daemon=True).start()

        # Return both the image and the count summary
        # Return both the image and the count summary with white background
        """return f"<div style='margin: 10px; padding: 10px; background-color: #FFFFFF; color: #000000; border-radius: 5px;'>" \
              f"{count_summary}" \
              f"<h3 style='color: black;'>Vehicle Detection Results</h3>" \
              f"<img src='data:image/jpeg;base64,{img_str}' style='max-width: 100%;'/></div>"


    except Exception as e:
        return f"Error during inference: {e}"""

        signal_html = simulate_traffic_signals(road_counts)

        return f"""
        <div style='display: grid; grid-template-columns: 1fr 1fr; gap: 20px;'>
            <div>{count_summary}</div>
            <div>{signal_html}</div>
        </div>
        <h3 style='color: black;'>Vehicle Detection Results</h3>
        <img src='data:image/jpeg;base64,{img_str}' style='max-width: 100%;'/>
        """

    except Exception as e:
        return f"Error during inference: {e}"

iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="filepath"),
    outputs="html",
    title="Road-wise Traffic Detection and Vehicle Counting",
    description="Upload a four-way junction image to detect vehicles and get road-wise vehicle counts by type. The system assumes vertical road width is 20% and horizontal road width is 25% of the image.",
    examples=[],
    cache_examples=False
)

iface.launch(debug=True, inline=False)
```
